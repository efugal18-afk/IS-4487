## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

```python
import matplotlib.pyplot as plt

transmission_counts = df['am'].value_counts()

transmission_labels = {0: 'Automatic', 1: 'Manual'}
labels = [transmission_labels[x] for x in transmission_counts.index]

fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(transmission_counts,
                                  labels=labels,
                                  autopct='%1.1f%%',

                                  startangle=90,
                                  pctdistance=0.85)

ax.set_title("Percentage of Automatic and Manual Transmissions")


ax.legend(wedges, labels,
          title="Transmission Type",
          loc="upper left",
          bbox_to_anchor=(-0.1, 1.),
          fontsize='small')

plt.tight_layout()
plt.show()
```

```python

import seaborn as sns

plt.figure(figsize=(15, 7)) #
sns.barplot(x='carmodel', y='hp', data=df, hue='carmodel', legend=False, palette='tab20')
plt.xticks(rotation=90)
plt.title("Horsepower (HP) by Car Model")
plt.xlabel("Car Model")
plt.ylabel("Horsepower (hp)")
plt.tight_layout()
plt.show()
```

```python
df.plot.scatter(x='wt', y='mpg', title='Weight vs. Miles Per Gallon')
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Miles/(US) gallon')
plt.show()
```

```python
import matplotlib.pyplot as plt

avg_mpg_by_cyl = df.groupby('cyl')['mpg'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(avg_mpg_by_cyl['cyl'], avg_mpg_by_cyl['mpg'], color='skyblue')
plt.title('Average MPG by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average Miles Per Gallon (MPG)')
plt.xticks(avg_mpg_by_cyl['cyl']) # Ensure all cylinder counts are shown as ticks
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

ax = df['wt'].plot(kind='hist', bins=10, title="Weight Distribution")
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Frequency')


for patch in ax.patches:
    x = patch.get_x() + patch.get_width() / 2
    y = patch.get_height()
    ax.annotate(f'{int(y)}', (x, y), ha='center', va='bottom')

plt.show()
```
