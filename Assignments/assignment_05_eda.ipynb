```python
plt.figure(figsize=(4,5))
sns.countplot(data=hotels, x='children', palette='pastel')
plt.xlabel('Number of Children')
plt.ylabel('Count of Bookings')
plt.title('Distribution of Children in Hotel Bookings')
plt.show()
```

```python
plt.figure(figsize=(3, 5))
sns.countplot(data=hotels, x='is_repeated_guest', palette='pastel')
plt.xlabel('Is Repeated Guest (0 = No, 1 = Yes)')
plt.ylabel('Count of Bookings')
plt.title('Distribution of Repeated Guests')
plt.xticks(ticks=[0, 1], labels=['New Guest', 'Repeated Guest'])
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.boxplot(y=hotels['lead_time'], palette='deep')
plt.ylabel('Lead Time (Days)')
plt.title('Distribution of Lead Time for Hotel Bookings')
plt.show()
```

## 5. Bivariate Analysis

### Business framing:  

Stakeholders often ask: ‚ÄúWhat drives cancellations?‚Äù or ‚ÄúDo longer stays mean higher revenue?‚Äù Bivariate analysis helps you uncover those kinds of relationships.

### Do the following:
- Choose 2 relevant variable pairs (e.g., `lead_time` vs. `is_canceled`, or `adr` vs. `customer_type`)

- Interpret what these relationships could mean for the hotel business

```python
# Your code to analyze variable relationships (e.g., scatterplots, grouped bars) üîß
plt.figure(figsize=(10, 10))
sns.regplot(data=hotels, x='days_in_waiting_list', y='lead_time', scatter_kws={'alpha':0.5})
plt.title('Lead Time vs. Days in Waiting List with Trendline')
plt.xlabel('Days in Waiting List')
plt.ylabel('Lead Time (Days)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.violinplot(data=hotels, x='is_repeated_guest', y='previous_cancellations', palette='dark')
plt.xlabel('Guest Type (0: New, 1: Repeated)')
plt.ylabel('Number of Previous Cancellations')
plt.title('Distribution of Previous Cancellations by Guest Type')
plt.xticks(ticks=[0, 1], labels=['New Guest', 'Repeated Guest'])
plt.show()
```
